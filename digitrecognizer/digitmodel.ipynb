```python
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
digit_trained = pd.read_csv('digit-recognizer/train.csv')
digit_tested = pd.read_csv('digit-recognizer/test.csv')
X = digit_trained.drop(columns = ['label'])
Y = digit_trained['label']
X_train , X_test , Y_train , Y_test = train_test_split(X,Y,test_size = 0.2)
model = DecisionTreeClassifier()
model.fit(X_train,Y_train)
prediction = model.predict(digit_tested)
prediction_df = pd.DataFrame({'ImageId':range(1,len(digit_tested)+1),'Label':prediction})
prediction_df.to_csv('label-submission.csv', index=False)



```
